#### Use your code from practical session 1 and create a function to perform GD for single variable linear regression.
#### The function should take the following input parameters:
##### Input data (X), Target Labels (y), Learning Rate (alpha), Maximum number of iterations
#### The function should return the following:
##### All Theta_0 and Theta_1 values till converge, and equivellant Losses, all hypothesis output, and optimum values of Theta_0 and Theta_1
#### Note: you can choose any suitable stop criteria.

```python
def Batch_GD(x_points,y_points,in_alfa,MaxIter):
    alfa = in_alfa
        
    y_actual = np.array(y_points)
    x_points = np.array(x_points)
    seta1 = 0
    seta0 = 0
    
    #assuming any value except befor using it in next while loop
    loss_diff_seta0 = 2
    loss_diff_seta1 = 2
    loss_norm = 10
    counter = 0
    j_list = []
    y_prdicted_list = []
    seta0_list = []
    seta1_list = []
    while abs(loss_norm) > .0001 and counter<MaxIter :
        y_prdicted =np.array([(seta1*x) +seta0 for x in x_points])
        y_prdicted_list.append(y_prdicted)
        seta0 = seta0 - (alfa*(1/y_actual.size)*sum((y_prdicted - y_actual)))
        seta0_list.append(seta0)
        seta1 = seta1 - (alfa*(1/y_actual.size)*sum((y_prdicted - y_actual)*x_points))
        seta1_list.append(seta1)
        
        loss = alfa*(1/(2*y_actual.size))*sum((y_prdicted - y_actual)**2)
        j_list.append(loss)
        
        loss_diff_seta0 = alfa*(1/y_actual.size)*sum((y_prdicted - y_actual))
        loss_diff_seta1 = alfa*(1/y_actual.size)*sum((y_prdicted - y_actual)*x_points)
        loss_norm = ((loss_diff_seta0**2)+(loss_diff_seta1**2))**.5

        print("end epoch",counter,seta0,seta1)
        counter +=1

    return seta0,seta1,y_prdicted,loss,j_list,y_prdicted_list,seta0_list,seta1_list
```

#### Try your function with any input parameterss of your choice.

```python
seta0_fi,seta1_fi,y_prdicted_fi,loss_fun,j_list,y_prdicted_list,seta0_list,seta1_list = Batch_GD(x,y,.01,5000)

print(seta0_fi,seta1_fi,y_prdicted_fi,loss_fun)
```

#### Update the implementation of the Batch GD to be Mini-Batch and rpeat the above plots and evaluations
#### Change batch size and see the effect

```python
def Mini_batch(x_points,y_points,in_alfa,MaxIter,batchSize):
    alfa = in_alfa
        
    y_actual = np.array(y_points)
    x_points = np.array(x_points)
    seta1 = 0
    seta0 = 0
    
    #assuming any value loss_diff except zero befor using it in next while loop
    loss_diff_seta0 = 2
    loss_diff_seta1 = 2
    loss_norm = 10
    counter = 0
    
    j_list = []
    y_prdicted_list = []
    seta0_list = []
    seta1_list = []
    if len(x_points) % batchSize != 0:
        num_of_batches = (len(x_points) // batchSize)+1
    else:
        num_of_batches = len(x_points) / batchSize
    
    print(num_of_batches)
    
    while abs(loss_norm) > .0001 and counter < MaxIter:
        size = 0
        y_epoch = []
        for point in range(int(num_of_batches)): 
            x_batch = x_points[size:(batchSize+size)]
            y_actual_batch  = y_actual[size:(batchSize+size)]
            
            y_prdicted =np.array([(seta1*x) +seta0 for x in x_batch])
            seta0 = seta0 - (alfa*(1/y_actual_batch.size)*sum((y_prdicted - y_actual_batch)))
            seta0_list.append(seta0)
            seta1 = seta1 - (alfa*(1/y_actual_batch.size)*sum((y_prdicted - y_actual_batch)*x_batch))
            seta1_list.append(seta1)

            loss = alfa*(1/(2*y_actual_batch.size))*sum((y_prdicted - y_actual_batch)**2)
            j_list.append(loss)

            loss_diff_seta0 = alfa*(1/y_actual_batch.size)*sum((y_prdicted - y_actual_batch))
            loss_diff_seta1 = alfa*(1/y_actual_batch.size)*sum((y_prdicted - y_actual_batch)*x_batch)
            loss_norm = ((loss_diff_seta0**2)+(loss_diff_seta1**2))**.5

            print("end iter",seta0,seta1)
            size = size + batchSize
            if size > len(x_points)-1 :
                size = len(x_points)-1 -size
                
            y_epoch = y_epoch + list(y_prdicted)
#             print(y_epoch)
            
        y_prdicted_list.append(y_epoch)
        counter+=1
    return seta0,seta1,y_epoch,loss,j_list,y_prdicted_list,seta0_list,seta1_list
```

```python
seta0_fi,seta1_fi,y_prdicted_fi,loss_fun,j_list,y_prdicted_list,seta0_list,seta1_list = Mini_batch(x,y,.001,5000,5)

print(seta0_fi,seta1_fi,y_prdicted_fi,loss_fun)
```

```python
r2_score(y, y_prdicted_fi)
```

## Stochastic GD

#### Update the implementation of the Batch GD to beStochastic and rpeat the above plots and evaluations
#### You can try different learning rate and see its effect.
#### Compare to Batch and Mini-Batch GD

```python
def SGD(x_points,y_points,in_alfa):
    alfa = in_alfa
        
    y_actual = np.array(y_points)
    x_points = np.array(x_points)
    seta1 = 0
    seta0 = 0
    
    #assuming any value except befor using it in next while loop
    loss_diff_seta0 = 2
    loss_diff_seta1 = 2
    loss_norm = 10
    j_list=[]
    seta0_list = []
    seta1_list = []
    y_prdicted_list = []
    
    while abs(loss_norm) > .0001:
        y_epoch = []
        for point in range(len(x_points)): 
            y_prdicted = (seta1*x_points[point]) +seta0 
            seta0 = seta0 - (alfa*(y_prdicted - y_actual[point]))
            seta0_list.append(seta0)
            seta1 = seta1 - (alfa*(y_prdicted - y_actual[point])*x_points[point])
            seta1_list.append(seta1)
            
            loss = alfa*(1/(2*y_actual.size))*(y_prdicted - y_actual[point])**2
            j_list.append(loss)
                          
            loss_diff_seta0 = alfa*(y_prdicted - y_actual[point])
            loss_diff_seta1 = alfa*(y_prdicted - y_actual[point])*x_points[point]
            loss_norm = ((loss_diff_seta0**2)+(loss_diff_seta1**2))**.5
            y_epoch.append(y_prdicted) 

            print("end iter",seta0,seta1)
        y_prdicted_list.append(y_epoch)
        
    y_prdicted_fi = (seta1*x_points) +seta0
    return seta0,seta1,y_prdicted_fi,loss,j_list,y_prdicted_list,seta0_list,seta1_list
```

```python
seta0_fi,seta1_fi,y_prdicted_fi,loss_fun,j_list,y_prdicted_list,seta0_list,seta1_list = SGD(x,y,.001)

print(seta0_fi,seta1_fi,y_prdicted_fi,loss_fun)
```
