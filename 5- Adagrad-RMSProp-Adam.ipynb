### Graphs Function
#### it's for making all required graphs in one function

```python
def GraphsFunction(j_list,seta0_list,seta1_list,y_prdicted_list,y_prdicted_fi):
    # Plot loss vs. epochs
    plt.plot(j_list)
    plt.xlabel('epochs')
    plt.ylabel('loss')
    plt.show()

    # Plot Theta_0 vs. loss 
    plt.plot(seta0_list,j_list);
    plt.xlabel('Theta_0')
    plt.ylabel('loss')
    plt.show()

    # Theta_1 vs. loss
    plt.plot(seta1_list,j_list);
    plt.xlabel('Theta_1')
    plt.ylabel('loss')
    plt.show()

    # Plot all regression lines till converge
    for y_prdictedGraph in y_prdicted_list:
        plt.plot(x,y_prdictedGraph);
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()

    # Plot the best regression line
    plt.plot(y_prdicted_fi)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
```

## Adagrad

### For a single variable linear regression ML model, build a function to find the optimum Theta_0 and Theta_1 parameters using Adagrad optimization algorithm.
#### The funtion should have the following input parameters:
##### 1. Input data as a matrix (or vector based on your data).
##### 2. Target label as a vector.
##### 3. Learning rate.
##### 4. Epsilon.
##### 5. Maximum number of iterations (Epochs).
#### The funtion should return the following outputs:
##### 1. All predicted Theta_0 in all iterations.
##### 2. All predicted Theta_1 in all iterations.
##### 3. Corresponding loss for each Theta_0 and Theta_1 predictions.
##### 4.All hypothesis outputs (prdicted labels) for each Theta_0 and Theta_1 predictions.
##### 5.Final Optimum values of Theta_0 and Theta_1.
#### Choose the suitable number of iterations, learning rate, Epsilon, and stop criteria.
#### Calculate r2 score. Shouldn't below 0.9
#### Plot the required curves (loss-epochs, loss-theta0, loss-theta1, all fitted lines per epoch (single graph), best fit line)
#### Try different values of the huperparameters and see the differnce in your results.

```python
def Adagrad(x_points,y_points,in_alfa,MaxIter,epslon):
    alfa = in_alfa
        
    y_actual = np.array(y_points)
    x_points = np.array(x_points)
    seta1 = 0
    seta0 = 0
    vt_seta0 = 0
    vt_seta1 = 0
    
    #assuming any value equal zero except loss_norm befor using it in next while loop
    loss_diff_seta0 = 2
    loss_diff_seta1 = 2
    loss_norm = 10
    counter = 0
    j_list = []
    y_prdicted_list = []
    seta0_list = []
    seta1_list = []
    while abs(loss_norm) > .00001 and counter<MaxIter :
        
        y_prdicted =np.array([(seta1*x) +seta0 for x in x_points])
        y_prdicted_list.append(y_prdicted)
        
        vt_seta0 = vt_seta0 + ((1/y_actual.size)*sum((y_prdicted - y_actual)))**2 
        seta0 = seta0 - ((1/y_actual.size)*sum((y_prdicted - y_actual)))*alfa / (vt_seta0**.5 + epslon)
        seta0_list.append(seta0)
        
        vt_seta1 = vt_seta1 + ((1/y_actual.size)*sum((y_prdicted - y_actual)*x_points))**2
        seta1 = seta1 - ((1/y_actual.size)*sum((y_prdicted - y_actual)*x_points))*alfa/ (vt_seta1**.5 + epslon)
        seta1_list.append(seta1)
        
        loss = (1/(2*y_actual.size))*sum((y_prdicted - y_actual)**2)
        j_list.append(loss)
        
        loss_diff_seta0 = (alfa/ (vt_seta1**.5 + epslon))*(1/y_actual.size)*sum((y_prdicted - y_actual))
        loss_diff_seta1 = (alfa/ (vt_seta1**.5 + epslon))*(1/y_actual.size)*sum((y_prdicted - y_actual)*x_points)
        loss_norm = ((loss_diff_seta0**2)+(loss_diff_seta1**2))**.5

        print("end epoch",counter,seta0,seta1)
        counter +=1

    return seta0,seta1,y_prdicted,loss,j_list,y_prdicted_list,seta0_list,seta1_list
```

```python
seta0_fi,seta1_fi,y_prdicted_fi,loss_fun,j_list,y_prdicted_list,seta0_list,seta1_list = Adagrad(x,y,.02,6000,1*10**-6)

print(seta0_fi,seta1_fi,y_prdicted_fi,loss_fun)
```

```python
# Calculate r2 score
r2_score(y, y_prdicted_fi)
```

## RMSProp

### Update the previos implementation to be RMSProp.
#### Compare your results with Adagrad results.

```python
def RMSProp(x_points,y_points,in_alfa,MaxIter,epslon,beta):
    alfa = in_alfa
    
    y_actual = np.array(y_points)
    x_points = np.array(x_points)
    seta1 = 0
    seta0 = 0
    vt_seta0 = 0
    vt_seta1 = 0
    
    #assuming any value equal zero except loss_norm befor using it in next while loop
    loss_diff_seta0 = 2
    loss_diff_seta1 = 2
    loss_norm = 10
    counter = 0
    j_list = []
    y_prdicted_list = []
    seta0_list = []
    seta1_list = []
    while abs(loss_norm) > .00001 and counter<MaxIter :
        
        y_prdicted =np.array([(seta1*x) +seta0 for x in x_points])
        y_prdicted_list.append(y_prdicted)
        
        vt_seta0 = (beta*vt_seta0) + ((1-beta)*((1/y_actual.size)*sum((y_prdicted - y_actual)))**2)
        seta0 = seta0 - ((1/y_actual.size)*sum((y_prdicted - y_actual)))*alfa / (vt_seta0**.5 + epslon)
        seta0_list.append(seta0)
        
        vt_seta1 = (beta*vt_seta1) + ((1-beta)*((1/y_actual.size)*sum((y_prdicted - y_actual)*x_points))**2)
        seta1 = seta1 - ((1/y_actual.size)*sum((y_prdicted - y_actual)*x_points))*alfa/ (vt_seta1**.5 + epslon)
        seta1_list.append(seta1)
        
        loss = (1/(2*y_actual.size))*sum((y_prdicted - y_actual)**2)
        j_list.append(loss)
        
        loss_diff_seta0 = (alfa/ (vt_seta0**.5 + epslon))*(1/y_actual.size)*sum((y_prdicted - y_actual))
        loss_diff_seta1 = (alfa/ (vt_seta1**.5 + epslon))*(1/y_actual.size)*sum((y_prdicted - y_actual)*x_points)
        loss_norm = ((loss_diff_seta0**2)+(loss_diff_seta1**2))**.5

        print("end epoch",counter,seta0,seta1)
        counter +=1

    return seta0,seta1,y_prdicted,loss,j_list,y_prdicted_list,seta0_list,seta1_list
```

## Adam

### Update the previos implementation to be Adam.
#### Compare your results with Adagrad and RMSProp results.

```python
def Adam(x_points,y_points,alfa,MaxIter,epslon,beta1,beta2):
    y_actual = np.array(y_points)
    x_points = np.array(x_points)
    seta1 = 0
    seta0 = 0
    vt_seta0 = 0
    vt_seta1 = 0
    mom_seta0 = 0
    mom_seta1 = 0
    
    #assuming any value equal zero except loss_norm befor using it in next while loop
    loss_diff_seta0 = 2
    loss_diff_seta1 = 2
    loss_norm = 10
    counter = 0
    j_list = []
    y_prdicted_list = []
    seta0_list = []
    seta1_list = []
    while abs(loss_norm) > .00001 and counter<MaxIter :
        
        y_prdicted =np.array([(seta1*x) +seta0 for x in x_points])
        y_prdicted_list.append(y_prdicted)
        
        vt_seta0 = (beta2*vt_seta0) + ((1-beta2)*((1/y_actual.size)*sum((y_prdicted - y_actual)))**2)
        mom_seta0 = (beta1*mom_seta0) + ((1-beta1)*((1/y_actual.size)*sum((y_prdicted - y_actual))))
        seta0 = seta0 - ((1/y_actual.size)*sum((y_prdicted - y_actual)))*(alfa / (vt_seta0**.5 + epslon))*mom_seta0
        seta0_list.append(seta0)
        
        vt_seta1 = (beta2*vt_seta1) + ((1-beta2)*((1/y_actual.size)*sum((y_prdicted - y_actual)*x_points))**2)
        mom_seta1 = (beta1*mom_seta0) + ((1-beta1)*((1/y_actual.size)*sum((y_prdicted - y_actual)*x_points)))
        seta1 = seta1 - ((1/y_actual.size)*sum((y_prdicted - y_actual)*x_points))*(alfa/ (vt_seta1**.5 + epslon))*mom_seta1
        seta1_list.append(seta1)
        
        loss = (1/(2*y_actual.size))*sum((y_prdicted - y_actual)**2)
        j_list.append(loss)
        
        loss_diff_seta0 = (alfa/ (vt_seta0**.5 + epslon))*(1/y_actual.size)*sum((y_prdicted - y_actual))
        loss_diff_seta1 = (alfa/ (vt_seta1**.5 + epslon))*(1/y_actual.size)*sum((y_prdicted - y_actual)*x_points)
        loss_norm = ((loss_diff_seta0**2)+(loss_diff_seta1**2))**.5

        print("end epoch",counter,seta0,seta1)
        counter +=1

    return seta0,seta1,y_prdicted,loss,j_list,y_prdicted_list,seta0_list,seta1_list
```

```python
seta0_fi,seta1_fi,y_prdicted_fi,loss_fun,j_list,y_prdicted_list,seta0_list,seta1_list = Adam(x,y,.02,1000,1*10**-8,.999,.95)

print(seta0_fi,seta1_fi,y_prdicted_fi,loss_fun)
```
